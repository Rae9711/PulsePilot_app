# Baseline definitions — quick example

This notebook demonstrates the `compute_average_post_metric` prototype from `src/metrics/baseline.py` and includes a small visualization example.

```python
# Ensure notebook can import local `src` package when opened from the `notebooks` folder
from pathlib import Path
import sys
sys.path.append(str(Path('..').resolve()))

from src.metrics.baseline import Feeling, compute_average_post_metric
from datetime import datetime

# Create synthetic feelings (post) over the last 14 days
now = datetime(2026, 3, 1)
feelings = [
    Feeling(occurred_at=datetime(2026, 2, 22), when='post', valence=4, energy=4, stress=2),
    Feeling(occurred_at=datetime(2026, 2, 24), when='post', valence=5, energy=5, stress=3),
    Feeling(occurred_at=datetime(2026, 2, 26), when='post', valence=4, energy=5, stress=2),
    Feeling(occurred_at=datetime(2026, 2, 28), when='post', valence=3, energy=3, stress=4),
    Feeling(occurred_at=datetime(2026, 3, 1), when='post', valence=5, energy=5, stress=1),
]

res = compute_average_post_metric(feelings, window_days=14, field='energy', now=now)
print('Metric:', res.metric)
print('Window (days):', res.window_days)
print('Value (avg energy):', res.value)
print('Data points:', res.data_points)
```

```python
# Visualization: plot energy over time for the synthetic feelings
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

dates = [f.occurred_at for f in feelings]
energies = [f.energy for f in feelings]

fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(dates, energies, marker='o', linestyle='-')
ax.set_title('Post-activity Energy over Time')
ax.set_xlabel('Date')
ax.set_ylabel('Energy (1-5)')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
fig.autofmt_xdate()
plt.grid(axis='y', alpha=0.3)
plt.show()
```
